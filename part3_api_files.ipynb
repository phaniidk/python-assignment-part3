```python
#TASK 2#:
import requests

# ==========================
# STEP 1: FETCH PRODUCTS
# ==========================

url = "https://dummyjson.com/products?limit=20"
response = requests.get(url)
data = response.json()["products"]

print(f"{'ID':<5} {'Title':<25} {'Category':<15} {'Price':<10} {'Rating'}")
print("-" * 70)

for p in data:
    print(f"{p['id']:<5} {p['title'][:24]:<25} {p['category']:<15} ${p['price']:<9} {p['rating']}")


# ==========================
# STEP 2: FILTER & SORT
# ==========================

filtered = [p for p in data if p["rating"] >= 4.5]
filtered.sort(key=lambda x: x["price"], reverse=True)

for p in filtered:
    print(f"{p['title']} | ${p['price']} | Rating: {p['rating']}")


# ==========================
# STEP 3: LAPTOP CATEGORY
# ==========================

url = "https://dummyjson.com/products/category/laptops"
response = requests.get(url)
laptops = response.json()["products"]

for p in laptops:
    print(p["title"], "-", "$" + str(p["price"]))


# ==========================
# STEP 4: POST REQUEST
# ==========================

payload = {
    "title": "My Custom Product",
    "price": 999,
    "category": "electronics",
    "description": "A product I created via API"
}

response = requests.post("https://dummyjson.com/products/add", json=payload)

print(response.json())
```

```python
#TASK 3#:
print("\n========== TASK 3A: SAFE DIVIDE ==========")

def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    except TypeError:
        return "Error: Invalid input types"

# Testing
print(safe_divide(10, 2))      # normal
print(safe_divide(10, 0))      # zero error
print(safe_divide("ten", 2))   # type error
```

```python
print("\n========== TASK 3B: FILE READER ==========")

def read_file_safe(filename):
    try:
        with open(filename, "r", encoding="utf-8") as f:
            return f.read()
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
    finally:
        print("File operation attempt complete.")

# Testing
print(read_file_safe("python_notes.txt"))   # should work
print(read_file_safe("ghost_file.txt"))     # should fail
```

```python
import requests

print("\n========== TASK 3C: ROBUST API ==========")

url = "https://dummyjson.com/products?limit=5"

try:
    response = requests.get(url, timeout=5)
    data = response.json()

    print("API Success!")
    for p in data["products"]:
        print(p["title"])

except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")

except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")

except Exception as e:
    print("Error:", e)
```

```python
print("\n========== TASK 3D: INPUT LOOP ==========")

import requests

# Simulated inputs (for Colab)
inputs = ["5", "101", "abc", "999", "quit"]

i = 0

while True:
    user = inputs[i]
    print("\nEnter product ID:", user)

    if user.lower() == "quit":
        break

    if not user.isdigit() or not (1 <= int(user) <= 100):
        print("Invalid input. Enter number between 1–100.")
        i += 1
        continue

    url = f"https://dummyjson.com/products/{user}"

    try:
        response = requests.get(url, timeout=5)

        if response.status_code == 404:
            print("Product not found.")
        else:
            data = response.json()
            print(data["title"], "-", "$" + str(data["price"]))

    except requests.exceptions.ConnectionError:
        print("Connection failed.")

    except requests.exceptions.Timeout:
        print("Timeout.")

    except Exception as e:
        print("Error:", e)

    i += 1
```

```python
from datetime import datetime
import requests

print("\n========== TASK 4: LOGGING ==========")

# Step 1: Logger function
def log_error(function_name, error_type, message):
    with open("error_log.txt", "a") as f:
        time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        f.write(f"[{time}] ERROR in {function_name}: {error_type} — {message}\n")


# Step 2: Trigger Connection Error
try:
    requests.get("https://this-host-does-not-exist-xyz.com/api", timeout=5)
except requests.exceptions.ConnectionError as e:
    log_error("fetch_products", "ConnectionError", str(e))
    print("Connection error logged.")


# Step 3: Trigger HTTP 404
url = "https://dummyjson.com/products/999"
response = requests.get(url)

if response.status_code != 200:
    log_error("lookup_product", "HTTPError", f"{response.status_code} Not Found for product ID 999")
    print("HTTP error logged.")


# Step 4: Read log file
print("\n========== ERROR LOG CONTENT ==========")

with open("error_log.txt", "r") as f:
    print(f.read())
```
